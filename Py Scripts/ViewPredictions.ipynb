```
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```
os.chdir('/content/')
```

```
files = sorted([i for i in os.listdir() if i.split('.')[1] == 'csv'])
for i, file in enumerate(files):
  print(i, file)
file_number = int(input('From the above list, enter the index of Stock whose predictions you want to view:\n'))

curr_file = open(files[file_number], 'r', newline='')
```

```
df = pd.read_csv(curr_file)

df
```

### LSTM

```
plt.figure(figsize=(12, 8))
plt.plot(df['Real Price'], color='red', label='Real Prices')
plt.plot(df['LSTM'], color='blue', label='Predicted Prices(LSTM)')
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
